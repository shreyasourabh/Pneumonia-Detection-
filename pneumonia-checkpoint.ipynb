```python
import os
import numpy as np
import torch
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
%matplotlib inline  
import gc 
```

```python
train_on_gpu = torch.cuda.is_available()
if not train_on_gpu:
    print('Cuda is not available')
else:
    print('Cuda is available. Training on GPU')
```

```python
data_dir=r'/content/drive/My Drive/chest_xray'
train_dir =os.path.join(data_dir,'train/')
val_dir =os.path.join(data_dir,'val/')
test_dir =os.path.join(data_dir,'test/')
```

```python
classes = ['Normal', 'Pneumonia']
```

```python
#Transforming the data
data_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
train_data =datasets.ImageFolder(train_dir,transform=data_transform)
valid_data =datasets.ImageFolder(val_dir,transform=data_transform)
test_data =datasets.ImageFolder(test_dir,transform=data_transform)
```

```python
batch_size = 32
num_workers =0
train_loader = torch.utils.data.DataLoader(train_data,batch_size= batch_size, num_workers=num_workers, shuffle=True)
val_loader = torch.utils.data.DataLoader(valid_data,batch_size= batch_size, num_workers=num_workers, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data,batch_size= batch_size, num_workers=num_workers, shuffle=True)
```

```python
#visualize some sample images
#obtain 1 batch of training image
dataiter = iter(train_loader)
images,labels =dataiter.next()
images =images.numpy()
```

```python
fig = plt.figure(figsize=(25,4))
for idx in np.arange(20):
    ax = fig.add_subplot(2,20/2,idx+1,xticks=[],yticks =[])
    plt.imshow(np.transpose(images[idx],(1,2,0)))
    ax.set_title(classes[labels[idx]])
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.conv1 = nn.Conv2d(3,16,3,padding=1)
        self.conv2 = nn.Conv2d(16,32,3,padding=1)
        #self.conv2_bn = nn.BatchNorm2d(32)
        self.conv3 = nn.Conv2d(32,64,3,padding=1)
        #self.conv3_bn = nn.BatchNorm2d(64)
        self.pool = nn.MaxPool2d(2,2)
        self.fc1 = nn.Linear(64*28*28,500)
        self.fc2 = nn.Linear(500,2)
        self.dropout = nn.Dropout(0.3)
    
    def forward(self,x):
        #x = torch.randn(16, 3, 3, 3)
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))

        #flattern the image
        #print(x.shape)
        x = x.view(-1,64*28*28)
        x = self.dropout(x)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        return x
    
model = Net()
print(model)

if train_on_gpu:
    model.cuda()
```

```python
import torch.optim as optim
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
gc.collect() 
```

```python
n_epochs = 30
valid_loss_min = np.Inf
for epoch in range(1, n_epochs+1):
    train_loss = 0.0
    valid_loss =0.0
    train = []
    val = []
    model.train()
    for data, target in train_loader:
        if train_on_gpu :
            data,target =data.cuda(), target.cuda()
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        train_loss+=loss.item() * data.size(0)

    model.eval()
    for data,target in val_loader:
        if train_on_gpu:
            data,target =data.cuda(), target.cuda()
        output = model(data)
        loss = criterion(output, target)
        valid_loss+=loss.item()*data.size(0)
    
    train_loss = train_loss/len(train_loader.sampler)
    train.append(train_loss)
    valid_loss = valid_loss/len(val_loader.sampler)
    val.append(valid_loss)
    print('Epoch : {} \t Training Loss : {:6f} \t Validation Loss : {:6f}'.format (epoch,train_loss,valid_loss))
    
    if valid_loss<=valid_loss_min : 
        print('Validation loss decreased ({:.6f} -> {:.6f}). Saving Model...'.format(valid_loss_min,valid_loss))
        torch.save(model.state_dict(),'model_tuber.pt')
        valid_loss_min =valid_loss
        
    model.load_state_dict(torch.load('model_tuber.pt'))
```

```python
nb_classes = 2
total = 0
correct =0
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
confusion_matrix = torch.zeros(nb_classes, nb_classes)
with torch.no_grad():
    for i, (inputs, classes) in enumerate(test_loader):
        inputs = inputs.to(device)
        classes = classes.to(device)
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)

        for t, p in zip(classes.view(-1), preds.view(-1)):
                confusion_matrix[t.long(), p.long()] += 1

print(confusion_matrix.numpy())

from pytorch_lightning.metrics.functional import accuracy
acc = accuracy(preds, classes).cpu()

print('Test accuracy: {:.3f}\n'.format(acc.numpy()))
```
